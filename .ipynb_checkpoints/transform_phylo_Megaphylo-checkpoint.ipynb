```python
import pandas as pd
from tqdm import tqdm
from ete3 import Tree
```

```python
def add_dist_to_root(tree):
    """
    Adds 'dist_to_root' attribute to each node, representing the distance from the root to the node.
    """
    for node in tree.traverse("preorder"):
        if node.is_root():
            node.dist_to_root = 0
        else:
            node.dist_to_root = node.up.dist_to_root + node.dist
```

```python
def add_dist_to_present(tree):
    """
    Adds 'dist_to_present' attribute to each node.
    For leaves, sets 'dist_to_present' to 0.
    For internal nodes, computes the distance from the node to the deepest tip.
    """
    max_depth = max(leaf.dist_to_root for leaf in tree.iter_leaves())
    for node in tree.traverse():
        if node.is_leaf():
            node.dist_to_present = 0
        else:
            node.dist_to_present = max_depth - node.dist_to_root
```

```python
with open('../CSIC-PhyloAI/data/simulations/87_t_params_SR.csv', "w") as f:
    f.write("index,birth_rate,turnover_rate,extinction_rate,diversification_rate,sampling_frac,tree_size,tree_path\n")
    
for i, row in tqdm(df.iterrows(), total=len(df)):
    tr = Tree(row['tree'])
    add_dist_to_root(tr)
    add_dist_to_present(tr)
    
    a0 = row['a0']
    a1 = row['a1']
    r0 = row['r0']
    r1 = row['r1']
    br0 = r0/(1-a0) 
    br1 = r1/(1-a1)
    er0 = br0-r0 
    er1 = br1-r1
    
    for n in tr.traverse():
        if n.dist_to_present < row['time']:
            n.birth_rate= br0
            n.extinction_rate = er0
            n.turnover_rate = a0
            n.diversification_rate = r0
        else:
            n.birth_rate= br1
            n.extinction_rate = er1
            n.turnover_rate = a1
            n.diversification_rate = r1
            
    out_tr = tr.write(
        features=['birth_rate', 'extinction_rate', 'turnover_rate', 'diversification_rate'],
        format_root_node=True, format=3)
    
    tr_path = 'data/simulations/TreeSim/87_t/BD_trees_SR/tree_' + str(i) + '.nwk'
    file_path = '../CSIC-PhyloAI/' + tr_path
    with open(file_path, "w") as f:
        f.write(out_tr)
        
    with open('../CSIC-PhyloAI/data/simulations/87_t_params_SR.csv', "a") as f:
        f.write(f"{i},\"[{br1}, {br0}]\",\"[{a1}, {a0}]\",\"[{er1}, {er0}]\",\"[{r1}, {r0}]\",{row['frac0']},{len(tr)},{tr_path}\n")
```

```python
lambda0_list = []
mu0_list = []
lambda1_list = []
mu1_list = []

for i, row in df.iterrows():
    # Segment 0
    r0 = row["r0"]
    a0 = row["a0"]
    lam0 = r0 / (1 - a0)
    mu0  = lam0 - r0
    
    lambda0_list.append(lam0)
    mu0_list.append(mu0)

    # Segment 1
    r1 = row["r1"]
    a1 = row["a1"]
    lam1 = r1 / (1 - a1)
    mu1  = lam1 - r1
    
    lambda1_list.append(lam1)
    mu1_list.append(mu1)

```

```python
import numpy as np

lower_lambda = [min(lambda0_list), min(lambda1_list)]
upper_lambda = [max(lambda0_list), max(lambda1_list)]

lower_mu = [min(mu0_list), min(mu1_list)]
upper_mu = [max(mu0_list), max(mu1_list)]

print(lower_lambda, upper_lambda)
print(lower_mu, upper_mu)
```
