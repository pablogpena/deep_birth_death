```python
import pickle
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

from dataset_code.utilities_vec import load_trees_from_array, plot_latent_points
```

```python
pickle_base = '/workspace/deep_birth_death/deep_learning/pickles/simulations/dataset_'
empiric_path =  '/workspace/deep_birth_death/empirical/'
n_tips = ['674', '489', '87']
div_scenarios = ["BD", "HE", "ME","SAT", "SR", "WW"]
```

```python
data = dict()
for i in n_tips:
    with open(pickle_base + i + "_10k.pkl", 'rb') as f:
        data[i] = pickle.load(f)
```

```python
for tip in data: 
    #Load empiricals
    if tip == "674": 
        tree_vec= load_trees_from_array([empiric_path + "eucalypts.nwk"])
    if tip == "489":
        tree_vec= load_trees_from_array([empiric_path + "conifers.nwk"], return_resc_factor=False)
    if tip == "87":  
        tree_vec= load_trees_from_array([empiric_path + "cetaceans.nwk"], return_resc_factor=False)
    #PCA 
    print("*"*5 + " " + tip + " " + "*"*5)
    pca = PCA(n_components=2, random_state=123)
    embedding_pca = pca.fit_transform(data[tip]['X_train'])
    empirical_pca = pca.transform(tree_vec)
    plot_latent_points(embedding_pca, np.argmax(data[tip]["y_class_train"], axis=1), div_scenarios,
                   empirical=empirical_pca)
    explained_variance_ratio = pca.explained_variance_ratio_
    print("Explained variance for axis 1: " + str(explained_variance_ratio[0]))
    print("Explained variance for axis 2: " + str(explained_variance_ratio[1]))
```
